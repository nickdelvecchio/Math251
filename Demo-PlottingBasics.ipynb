```python
# Enable inline plotting
# This will cause your plots to be AUTO-RENDERED IN THE NOTEBOOK.
# We will discuss how to save plots at the end
%matplotlib inline
```

```python
# Basic plot
# (b) : Plotting
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-5,5,30)  # create 30 points from -5 to 5, uniformly spaced
y = x*x - 2*x
plt.plot(x,y)
```

```python
# Axis labels and titles
plt.plot(x)
plt.xlabel("hello")
plt.ylabel("there")
plt.title("title")
```

```python
# Legends
plt.plot(x,label='a')
plt.plot(x+1,label='b')
plt.plot(x+2, label='c')
plt.legend()
```

```python
# Latex in plot labels, legends, etc
#   - Beware 'escape characters' ... use 'r' at the front
plt.plot(x,label=r'$\sin(x)$')
plt.xlabel(r'$\xi$')
plt.ylabel(r'$P(< \xi^\alpha(\gamma))$')
plt.title("$\max \Gamma$")
plt.legend()
```
